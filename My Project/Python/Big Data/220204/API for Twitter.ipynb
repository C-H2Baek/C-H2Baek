2. ì¶”ì¶œ : í‚¤ì›Œë“œ ì¶”ì¶œ
[ë±ìŠ¤íŠ¸ ë°ì´í„° ì „ì²˜ë¦¬]
- í•œê¸€ ë¬¸ìì—´ì„ ê¸°ì¤€ìœ¼ë¡œ í‚¤ì›Œë“œ ì¶”ì¶œ
- tweet_textë°ì´í„°ì—ì„œ í•œê¸€ ë¬¸ìì—´ ë§Œì„ ì¶”ì¶œí•˜ì—¬ ko_textìƒì„±

```python
import re

# í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜ : í•œê¸€ ë° ë„ì–´ì“°ê¸° ì´ì™¸ì˜ ë¬¸ìëŠ” ì œê±°
def text_cleaning(text):
    # í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë¥¼ ì œì™¸í•œ ëª¨ë“  ê¸€ì íŒ¨í„´ì„ ì§€ì •í•˜ì—¬ hangulë¡œ ì •ì˜
    hangul = re.compile('[^ ã„±-ã…£ ê°€-í£]+')
    result = hangul.sub('', text) # í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë¥¼ ì œì™¸í•œ ëª¨ë“  ê¸€ì íŒ¨í„´ì„ '' ë¹ˆ ë¬¸ìë¡œ ì¹˜í™˜
    return result
def hangul_cleaning(text):
    nohangul = re.compile('[ ã„±-ã…£ ê°€-í£]+')
    result2 = nohangul.sub(' ', text) # í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë¥¼ í¬í•¨í•œ ëª¨ë“  ê¸€ì íŒ¨í„´ì„ '' ë¹ˆ ë¬¸ìë¡œ ì¹˜í™˜
    return result2

# í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜text_cleaning()ë¥¼ 'tweet_text' í”¼ì²˜ì— ì´ë¥¼ ì ìš©, í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë§Œ ì¶”ì¶œí•˜ì—¬ ko_text í”¼ì²˜ ìƒì„±
df['ko_text'] = df['tweet_text'].apply(lambda x: text_cleaning(x))
df
```

### [konlpyë¥¼ í™œìš©í•œ ë¶ˆìš©ì–´ ì œê±°]

```python
from konlpy.tag import Okt # ëª…ì‚¬
from collections import Counter #ì¹´ìš´í„°

korean_stopwords_path = "../Library/korean_stopwords.txt"
nouns_tagger = Okt()
with open(korean_stopwords_path, encoding='utf-8') as f: # ë¶ˆìš©ì–´ í…ìŠ¤íŠ¸ íŒŒì¼ì„ ì—´ì–´ fë¡œ ì¹˜í™˜
    stopwords = f.readlines() # íŒŒì¼ë¡œë¶€í„° ë¶ˆìš©ì–´ë¥¼ í•œ ì¤„ì”© read
stopwords = [x.strip() for x in stopwords] # ì¤„ë°”ê¿ˆ ê³µë°± ì œê±° í›„ ë¦¬ìŠ¤íŠ¸ ìƒì„± 
more_stopwords = ['ì£¼ì‹','ê°œë‹˜', 'ë¹¡ì„¼', 'ì§„ì§œ', 'ë•Œë¬¸', keyword]
for stopword in more_stopwords:
    stopwords.append(stopword) # ì¶”ê°€ ë¶ˆìš©ì–´ ìŠ¤íƒ‘ì›Œë“œì— ì¶”ê°€
    
# í’ˆì‚¬ì¤‘ ëª…ì‚¬ë§Œ ì¶”ì¶œ ë° í•œê¸€ì ì œê±°
def get_nouns(x):
    nouns_tagger = Okt() # Okt() í´ë˜ìŠ¤ ì„ ì–¸
    nouns = nouns_tagger.nouns(x)
    
    # í•œê¸€ì í‚¤ì›Œë“œ ì œê±°
    nouns = [noun for noun in nouns if len(noun) > 1]
    # ë¶ˆìš©ì–´ë¥¼ ì œê±°
    nouns = [noun for noun in nouns if noun not in stopwords]
    return nouns

df['nouns'] = df['ko_text'].apply(lambda x: get_nouns(x))
df
```

```python
# ì—°ê´€ ë¶„ì„ ì—°ìŠµ
from apyori import apriori

transactions = [
    ['game', 'diablo']
]

# ì—°ê´€ ë¶„ì„ì„ ìˆ˜í–‰
results= list(apriori(transactions))
for result in results:
    print(result)
```
