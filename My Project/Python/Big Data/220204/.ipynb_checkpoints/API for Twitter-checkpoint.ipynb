2. ì¶”ì¶œ : í‚¤ì›Œë“œ ì¶”ì¶œ
[ë±ìŠ¤íŠ¸ ë°ì´í„° ì „ì²˜ë¦¬]
- í•œê¸€ ë¬¸ìì—´ì„ ê¸°ì¤€ìœ¼ë¡œ í‚¤ì›Œë“œ ì¶”ì¶œ
- tweet_textë°ì´í„°ì—ì„œ í•œê¸€ ë¬¸ìì—´ ë§Œì„ ì¶”ì¶œí•˜ì—¬ ko_textìƒì„±

```python
import re

# í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜ : í•œê¸€ ë° ë„ì–´ì“°ê¸° ì´ì™¸ì˜ ë¬¸ìëŠ” ì œê±°
def text_cleaning(text):
    # í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë¥¼ ì œì™¸í•œ ëª¨ë“  ê¸€ì íŒ¨í„´ì„ ì§€ì •í•˜ì—¬ hangulë¡œ ì •ì˜
    hangul = re.compile('[^ ã„±-ã…£ ê°€-í£]+')
    result = hangul.sub('', text) # í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë¥¼ ì œì™¸í•œ ëª¨ë“  ê¸€ì íŒ¨í„´ì„ '' ë¹ˆ ë¬¸ìë¡œ ì¹˜í™˜
    return result
def hangul_cleaning(text):
    nohangul = re.compile('[ ã„±-ã…£ ê°€-í£]+')
    result2 = nohangul.sub(' ', text) # í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë¥¼ í¬í•¨í•œ ëª¨ë“  ê¸€ì íŒ¨í„´ì„ '' ë¹ˆ ë¬¸ìë¡œ ì¹˜í™˜
    return result2

# í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜text_cleaning()ë¥¼ 'tweet_text' í”¼ì²˜ì— ì´ë¥¼ ì ìš©, í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë§Œ ì¶”ì¶œí•˜ì—¬ ko_text í”¼ì²˜ ìƒì„±
df['ko_text'] = df['tweet_text'].apply(lambda x: text_cleaning(x))
df
```

### [konlpyë¥¼ í™œìš©í•œ ë¶ˆìš©ì–´ ì œê±°]

```python
from konlpy.tag import Okt # ëª…ì‚¬
from collections import Counter #ì¹´ìš´í„°

korean_stopwords_path = "../Library/korean_stopwords.txt"
nouns_tagger = Okt()
with open(korean_stopwords_path, encoding='utf-8') as f: # ë¶ˆìš©ì–´ í…ìŠ¤íŠ¸ íŒŒì¼ì„ ì—´ì–´ fë¡œ ì¹˜í™˜
    stopwords = f.readlines() # íŒŒì¼ë¡œë¶€í„° ë¶ˆìš©ì–´ë¥¼ í•œ ì¤„ì”© read
stopwords = [x.strip() for x in stopwords] # ì¤„ë°”ê¿ˆ ê³µë°± ì œê±° í›„ ë¦¬ìŠ¤íŠ¸ ìƒì„± 
more_stopwords = ['ì£¼ì‹','ê°œë‹˜', 'ë¹¡ì„¼', 'ì§„ì§œ', 'ë•Œë¬¸', keyword]
for stopword in more_stopwords:
    stopwords.append(stopword) # ì¶”ê°€ ë¶ˆìš©ì–´ ìŠ¤íƒ‘ì›Œë“œì— ì¶”ê°€
    
# í’ˆì‚¬ì¤‘ ëª…ì‚¬ë§Œ ì¶”ì¶œ ë° í•œê¸€ì ì œê±°
def get_nouns(x):
    nouns_tagger = Okt() # Okt() í´ë˜ìŠ¤ ì„ ì–¸
    nouns = nouns_tagger.nouns(x)
    
    # í•œê¸€ì í‚¤ì›Œë“œ ì œê±°
    nouns = [noun for noun in nouns if len(noun) > 1]
    # ë¶ˆìš©ì–´ë¥¼ ì œê±°
    nouns = [noun for noun in nouns if noun not in stopwords]
    return nouns

df['nouns'] = df['ko_text'].apply(lambda x: get_nouns(x))
df
```

```python
# ì—°ê´€ ë¶„ì„ ì—°ìŠµ
from apyori import apriori

transactions = [
    ['game', 'diablo'],
    ['ì†í¥ë¯¼', 'ì‹œì†Œì½”'],
    ['ì†í¥ë¯¼', 'ì¼€ì¸'],
    ['ì†í¥ë¯¼', 'ì¼€ì¸', 'í¬ì²´í‹°ë…¸']
]

# ì—°ê´€ ë¶„ì„ì„ ìˆ˜í–‰
results= list(apriori(transactions))
for result in results:
    print(result)
```

```python
# ì§€ì§€ 0.5 ì‹ ë¢° 0.6 í–¥ìƒ 1.0 ì´ìƒ ê·œì¹™ì˜ í¬ê¸° 2ì´í•˜ ì¶”ì¶œ
list(apriori(transactions,
            min_support=0.5,
            min_confidence=0.6,
            min_lift=1.0,
            max_legnth=2))
```

### íŠ¸ìœ„í„° ì—°ê´€ í‚¤ì›Œë“œ ë¶„ì„

```python
# íŠ¸ëœì ì…˜ ë°ì´í„°ë¥¼ ì¶”ì¶œ
transactions = df['nouns'].tolist()
print(transactions)
```

```python
# ì—°ê´€ ë¶„ì„ì„ ìˆ˜í–‰
results = list(apriori(transactions,
                       min_support=0.105,
                       min_confidence=0.1,
                       min_lift=5,
                       max_length=2))
print(results)
```

```python
# ë°ì´í„° í”„ë ˆì„ í˜•íƒœë¡œ ì •ë¦¬
# ë„¤íŠ¸ì› ê´€ê³„ ì„ 
columns = ['source', 'target', 'support']
network_df = pd.DataFrame(columns=columns)

# ê·œì¹™ì˜ ì¡°ê±´ì ˆì„ source, ê²°ê³¼ì ˆì„ target, ì§€ì§€ë„ë¥¼ support ë¼ëŠ” ë°ì´í„° í”„ë ˆì„ì˜ í”¼ì²˜ë¡œ ë³€í™˜
for result in results:
    if len(result.items) == 2:
        items = [x for x in result.items]
        row = [items[0], items[1], result.support]
        series = pd.Series(row, index=network_df.columns)
        network_df = network_df.append(series, ignore_index=True)

network_df.head()
```

```python
# ë§ë­‰ì¹˜ë¥¼ ì¶”ì¶œí•©
tweet_corpus = "".join(df['ko_text'].tolist())
print(tweet_corpus)
```

```python
from konlpy.tag import Okt
from collections import Counter

# ëª…ì‚¬ í‚¤ì›Œë“œë¥¼ ì¶”ì¶œ, ë¹ˆë„ìˆ˜ ê³„ì‚°
nouns_tagger = Okt()
nouns = nouns_tagger.nouns(tweet_corpus)
count = Counter(nouns)

# í•œê¸€ì í‚¤ì›Œë“œë¥¼ ì œê±°
remove_char_counter = Counter({x: count[x] for x in count if len(x) > 1})
remove_char_counter = Counter({x: remove_char_counter[x] for x in count if x not in stopwords})
print(remove_char_counter)
```

```python
# í‚¤ì›Œë“œì™€ í‚¤ì›Œë“œ ë¹ˆë„ ì ìˆ˜ë¥¼ â€˜nodeâ€™, â€˜nodesizeâ€™ ë¼ëŠ” ë°ì´í„° í”„ë ˆì„ì˜ í”¼ì²˜ë¡œ ìƒì„±
node_df = pd.DataFrame(remove_char_counter.items(), columns=['node', 'nodesize'])
node_df = node_df[node_df['nodesize'] >= 50] # ì‹œê°í™”ì˜ í¸ì˜ë¥¼ ìœ„í•´ â€˜nodesizeâ€™ 50 ì´í•˜ëŠ” ì œê±°í•©ë‹ˆë‹¤.
node_df.head()
```

# <Step4. ì‹œê°í™”> : ì—°ê´€ í‚¤ì›Œë“œ ë„¤íŠ¸ì›Œí¬ ì‹œê°í™”

### [ì—°ê´€ í‚¤ì›Œë“œ ë„¤íŠ¸ì›Œí¬ ì‹œê°í™”]

- ì•„ë˜ ì½”ë“œ ì‹¤í–‰ì„ ìœ„í•´, anaconda prompt í˜¹ì€ Terminalì—ì„œ ì•„ë˜ì™€ ê°™ì€ íŒ¨í‚¤ì§€ë¥¼ ì„¤ì¹˜í•´ ì¤ë‹ˆë‹¤.
    - (env_name) `pip install networkx`
- í˜¹ì€ ì•„ë˜ì˜ ì½”ë“œë¡œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì„¤ì¹˜í•©ë‹ˆë‹¤.
- networkx ë²„ì „ 2.3 ì´í•˜ë¥¼ ì„¤ì¹˜í•˜ëŠ” ê²½ìš°, ì•„ë˜ ì½”ë“œë¥¼ ë‹¤ìŒê³¼ ê°™ì´ ë°”ê¿”ì£¼ì–´ì•¼ í•©ë‹ˆë‹¤.
    - `sizes = [G.nodes[node]['nodesize']*25 for node in G]` -> `sizes = [G.node[node]['nodesize']*25 for node in G]`

```python
import networkx as nx
plt.figure(figsize=(15,15))

# networkx ê·¸ë˜í”„ ê°ì²´ë¥¼ ìƒì„±
G = nx.Graph()

# node_dfì˜ í‚¤ì›Œë“œ ë¹ˆë„ìˆ˜ë¥¼ ë°ì´í„°ë¡œ í•˜ì—¬, ë„¤íŠ¸ì›Œí¬ ê·¸ë˜í”„ì˜ â€˜ë…¸ë“œâ€™ ì—­í• ì„ í•˜ëŠ” ì›ì„ ìƒì„±
for index, row in node_df.iterrows():
    G.add_node(row['node'], nodesize=row['nodesize'])
    
# network_dfì˜ ì—°ê´€ ë¶„ì„ ë°ì´í„°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ, ë„¤íŠ¸ì›Œí¬ ê·¸ë˜í”„ì˜ â€˜ê´€ê³„â€™ ì—­í• ì„ í•˜ëŠ” ì„ ì„ ìƒì„±
for index, row in network_df.iterrows():
    G.add_weighted_edges_from([(row['source'], row['target'], row['support'])])
    
# ê·¸ë˜í”„ ë””ìì¸ê³¼ ê´€ë ¨ëœ íŒŒë¼ë¯¸í„°ë¥¼ ì„¤ì •
pos = nx.random_layout(G)
#pos = nx.spring_layout(G, k=0.6, iterations=50)
sizes = [G.nodes[node]['nodesize']*25 for node in G]
nx.draw(G, pos, node_size=sizes, node_color ='pink' )
nx.draw_networkx_labels(G, pos, font_family='HYSUPM', font_size=20)

# ê·¸ë˜í”„ë¥¼ ì¶œë ¥
ax = plt.gca()
plt.show()
```
