# <Step 2> 인사이트의 발견 : 탐색과 시각화하기

두 피처 간의 상관 계수 구하기

```python
# beer_servings, wine_servings 두 피처 간의 상관 계수를 계산.
# perason은 상관 계수를 구하는 계산 방법 중 하나를 의미하며, 가장 널리 쓰이는 방법.
corr = drinks[['beer_servings', 'wine_servings']].corr(method = 'pearson')
corrs = drinks.corr(method = 'pearson')
corr
```

```python
corrs
```

#히트맵(heatmap), 페어플롯(pairplot) 기법 사용하기

 히트맵(heatmap) : 히트(head)와 시도를 뜻하는 맵(map)을 결합시킨 단어로 색상으로 표현할 수 있는 다양한 정보를 일정한 이미지 위헤 열분포 형태의 그래픽으로 출력

여러 피처의 상관 관계 분석하기

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
sns.set(font_scale=1.2)

# 그래프 출력을 위한 cols 이름
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

# corr행렬 히트맵을 시각화
# 히트맵(heatmap)을 통해 corr.values 를 파라미터로 넣어줌
hm = sns.heatmap(corrs.values,
                cbar=True,
                cbar_kws={"orientation": "horizontal"},
                annot=True,
                square=True,
                fmt='.2f',
                annot_kws={'size': 15},
                linewidths=2,
                cmap="Spectral",
                yticklabels=cols_view,
                xticklabels=cols_view)

plt.tight_layout()
plt.show
```

```python
# 시각화 라이브러리를 이용한 피처간의 scatter plot을 출력
# 페어플롯(pairplot)은 데이터 프레임을 파라미터로 넣어 줌
sns.set(style='dark', palette="pastel" ,context='notebook')
sns.pairplot(drinks[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']], height = 2.5, kind="kde")
plt.show()
```

# 인사이트 도출1 : 대륙(continent)별 평균 wine_servings 탐색

```python
drinks.groupby('continent').wine_servings.mean()
```

# 인사이트 도출2 :전체 평균보다 적은 알코올을 섭취하는 대륙

step 1 . 대륙별 알코올 섭취량의 평균

```python
#알콜 섭취양 total_litres_of_pure_alcohol
AveCon = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()
AveCon
```

step 2 . 알콜 섭취량의 전체 평균

```python
AveAll = drinks.total_litres_of_pure_alcohol.mean()
AveAll
```

step 3 . 평균 이상의 알코올을 섭취하는 대륙

```python
AveAllOver_Result = AveCon[AveCon >= AveAll]
AveAllOver_Result
```

step 4 . 평균 이하의 알코올을 섭취하는 대륙

```python
AveAllLower_Result = AveCon[AveCon <= AveAll]
AveAllLower_Result
```

```python
AveAllOver = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()
AveAllOver_Result = AveAllOver[AveAllOver >= AveAll]
AveAllOver_Result
```

```python
AveAllLower = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()
AveAllLower_Result = AveAllLower[AveAllLower < AveAll].index.tolist()
AveAllLower_Result
```

```python
AveAllLower = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()
AveAllLower_Result = AveAllLower[AveAllLower < AveAll].values.tolist()
AveAllLower_Result
```
