```python
Lists = Soup.select('div.eview-box')
for L in Lists:
    Lecture = Soup.select_one('div.r-eview > a > em').text
    Title = L.select_one('span.font14.bold').text
    Code = L.select_one('p:nth-child(2)').text
    Date = L.select_one('p:nth-child(3)').text
    Time = L.select_one('p:nth-child(4)').text.strip()
    Cost = L.select_one('p:nth-child(5)').text
    Reward = L.select_one('p:nth-child(5) > span.bold.purple')
    print(f'\n과정구분 : {Lecture}\n과정명 : {Title}\n{Code}\n{Date}\n{Time}\n{Cost}\n{Reward}')
    print('-----------------------------------------------------------------------------------------------------------------------')
 
```

```python
Lists = Soup.select('div.eview-box')
for L in Lists:
    Reward = L.select_one('p:nth-child(5) > span.bold.purple')
    
    if Reward != None:
        Title = L.select_one('span.font14.bold').text
        Lecture = Soup.select_one('div.r-eview > a > em').text
        Code = L.select_one('p:nth-child(2)').text
        Date = L.select_one('p:nth-child(3)').text
        Time = L.select_one('p:nth-child(4)').text.strip()
        S = re.sub(r"[\t\n\"]*", "", Time)
        
        Cost = L.select_one('p:nth-child(5)').text
        print(f'과정구분 : {Lecture}\n과정명 : {Title}\n{Code}\n{Date}\n{S}\n{Cost}')
        print('-----------------------------------------------------------------------------------------------------------------------')
 
```
